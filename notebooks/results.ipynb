### Compounding factor

```python
users_groups_dict = {}
for fold_n in trange(5, desc='folds'):
    ds = DataSplitter(DATA_PATH, DEMO_PATH, out_dir=OUT_DIR)
    pandas_dir_path, _, _, _ = ds.get_paths(fold_n=fold_n)
    user_groups = ds.get_user_groups_indxs(pandas_dir_path, 'gender')
    for user_group in user_groups:
        
        users_groups_dict[('standard',user_group.name,fold_n)] = len(user_group.te_idxs)
        users_groups_dict[('up_sample',user_group.name,fold_n)] = len(user_group.te_idxs) # Test are the same
        
    #ds = DataSplitter(DOWN_DATA_PATH, DOWN_DEMO_PATH, out_dir=OUT_DIR)
    #pandas_dir_path, _, _, _ = ds.get_paths(fold_n=fold_n)
    #user_groups = ds.get_user_groups_indxs(pandas_dir_path, 'gender')
    #for user_group in user_groups:
    #   users_groups_dict[('down_sample',user_group.name,fold_n)] = len(user_group.te_idxs)
groups_counts= pd.Series(users_groups_dict).unstack()
groups_counts
```

```python
# Selecting the results for m and f
m_f_results  = results.unstack(-2)[['m','f']].stack().unstack(2)
m_f_results
```

```python
# Same indexes as groups_counts (need to place the same indexes back!)
m_f_results_tmp = m_f_results.swaplevel(0,1).swaplevel(1,-1).swaplevel(-1,-2).sort_index()
m_f_results_tmp
```

```python
# Reindexing groups_counts and performing a 1-to-1 multiplication
weighted_average = groups_counts.reindex(m_f_results_tmp.index,method='ffill').mul(m_f_results_tmp)
weighted_average
```

```python
# Summing values on the fold level 
summed = weighted_average.transpose().sum()
summed
```

```python
# Summing on m f level to normalize
total_sum = summed.unstack(1).transpose().sum()
total_sum
```

```python
# Lastly, normalizing the summed values by the total_sum! (we fill with 0 where the normalizing factor was 0)
compounding_factors = summed.unstack(1).div(total_sum,axis='index').fillna(0)
compounding_factors
```

```python
# Transforming group counts in percentages for the collection distribution
groups_percentages = groups_counts.sum(axis='columns').div(groups_counts.sum(axis='columns').sum(level=0),level=0)
groups_percentages
```

#### KL related value

```python
def kl_event_diff(p, q):
    return - (p.loc['m'] * np.log2(p.loc['m']/q.loc['m'])) + (p.loc['f'] * np.log2(p.loc['f']/q.loc['f']))
def kl_divergence(p, q):
    return (p.loc['m'] * np.log2(p.loc['m']/q.loc['m'])) + (p.loc['f'] * np.log2(p.loc['f']/q.loc['f']))
```

```python
# Same indexes as groups_percentages (need to place the same indexes back!)
compounding_factors_tmp = compounding_factors.stack().swaplevel(1,-1).sort_index()
compounding_factors_tmp
```

```python
# Reindexing groups_percentages
groups_percentages_tmp = groups_percentages.reindex_like(compounding_factors_tmp,method='ffill')
groups_percentages_tmp
```

```python
# Showing m and f in the axis
# RUN ONLY ONCE!
groups_percentages_tmp = groups_percentages_tmp.unstack(1).transpose()
compounding_factors_tmp = compounding_factors_tmp.unstack(1).transpose()
compounding_factors_tmp
```

```python
# Combining the two
kl_diffs = groups_percentages_tmp.combine(compounding_factors_tmp,overwrite=False,func=kl_event_diff)
# Removing m and f indexes (since it's the same)
kl_diffs = kl_diffs.drop_duplicates().stack([0,1,2]).droplevel(0)
# Placing metrics on the columns
kl_diffs = kl_diffs.unstack(1).swaplevel().sort_index()
kl_diffs
```
